```python
table = []

for i in kk:
    c = requests.get(f'http://pharmanlp-177020.uc.r.appspot.com/api/1/StudyHack/ae/subject/{i}/list')
    dic = c.json()
    data = pd.DataFrame(dic)
    for j in data['data']:
        table.append(j)
```

```python
table1_ae=pd.DataFrame(table)
table1_ae
```

```python
table2 = []

for p in fourth['data']:
    d = requests.get(f'http://pharmanlp-177020.uc.r.appspot.com/api/1/StudyHack/cm/subject/{p}/list')
    sec = d.json()
    data2 = pd.DataFrame(sec)
    for q in data2['data']:
        table2.append(q)
```
